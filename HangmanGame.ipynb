```python
#Create a Hangman Game

import random
from hangman_words import word_list
from hangman_art import stages, logo
print(logo)

# Randomly choose a word as the answer
chosen_word = random.choice(word_list)
# Player has 6 lives by default
lives = 6

# Create display blanks according to how many letters
display = []
for cha in range(len(chosen_word)):
    display.append("_")
print (display)
print (stages[6])

# Game status
end_of_game = False

while end_of_game == False:

    # Player's guesses of letters
    guess = input("Please guess one letter.").lower()

    if guess in display:
        print (f"You've already guessed {guess}.")

    if guess not in chosen_word:
        lives -= 1
        print(f"Guessed letter {guess} is not in the word. You lose a life.")
        print(stages[lives])

    print(f"\nLives remaining: {lives}")

    if lives == 0:
        print (f"You lost! Game over! The answer is {chosen_word}.")
        break

    # n is counter to represent the index of the chosen word
    n = 0
    for letters in chosen_word:
        if guess == letters:
            display[n] = guess
        n += 1
    print(display)

    # Change game status if the player guess all letters in the word
    if "_" not in display:
        end_of_game = True
        print ("You win!")

```
